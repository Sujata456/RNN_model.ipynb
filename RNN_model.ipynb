```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
x=[]
y=[]
for i in range(1,10):
  x.append([i, i+1, i+2])
  y.append(i+3)

x=np.array(x).reshape((len(x),3,1))
y= np.array(y)
```

```python
# step 3
# build the RNN model
model = Sequential()
model.add(SimpleRNN(10, activation='relu',input_shape=(3,1)))
model.add(Dense(1))
model.compile(optimizer ='adam' ,loss='mse')

# train model
model.fit(x,y,epochs=300 ,verbose=1)
```

```python
# step 4
# test the model with a new sequence
test_input = np.array(([10,11,12])).reshape((1,3,1))
predicted = model.predict(test_input, verbose=0)

print(f"Input Sequence:[10,11,12]")
print(f"Predicted Next Number:{round(float(predicted[0][0]),2)}")

```
